# Análise univariada.

```python
# Quais as 10 idades que mais aparecem no dataset?

d1 = ''' SELECT Age, COUNT(*) as QtAge
          FROM diabetes
          GROUP BY Age
          ORDER BY QtAge DESC
          LIMIT 10 '''

re_d1 = pd.read_sql_query(d1, conn)
re_d1
```

```python
# Qual a distribuição de gravidez entre os indivíduos?

d2 = ''' SELECT Pregnancies, COUNT(*) AS QtPreg, ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM diabetes)) AS Percentual
          FROM diabetes
          GROUP BY Pregnancies
          ORDER BY QtPreg DESC
          LIMIT 10'''

re_d2 = pd.read_sql_query(d2, conn)
re_d2
```

```python
# De acordo com os valores de Glicemia, qual a quantidade de pessoas com hipoglicemia, saudáveis, pré-diabéticos e diabéticos?

d3 = ''' SELECT sub1.Classificação, COUNT(*) AS QtClassif, ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM diabetes)) AS Percentual
          FROM (
            SELECT Glucose,
          CASE
            WHEN Glucose >= 0 AND Glucose <= 69 THEN 'Hipoglicêmico'
            WHEN Glucose >= 70 AND Glucose <= 99 THEN 'Saudável'
            WHEN Glucose >= 100 AND Glucose <= 125 THEN 'Pré-diabético'
            ELSE 'Diabético'
        END AS Classificação
        FROM diabetes
        )as sub1
      GROUP BY sub1.Classificação
      ORDER BY QtClassif DESC'''

re_d3 = pd.read_sql_query(d3, conn)
re_d3
```

```python
# De acordo com os valores de massa corporal (BMI), quantas pessoas são abaixo do peso, peso normal, acima do peso, obesidade 1, obesidade 2, obesidade 3?

d4 = ''' SELECT sub1.Classificação, COUNT(*) AS QtClassif, ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM diabetes)) AS Percentual
          FROM (
            SELECT BMI,
          CASE
            WHEN BMI >= 0 AND BMI <= 18.4 THEN 'Abaixo do peso'
            WHEN BMI >= 18.5 AND BMI <= 24.8 THEN 'Peso normal'
            WHEN BMI >= 24.9 AND BMI <= 29.8 THEN 'Acima do peso'
            WHEN BMI >= 29.9 AND BMI <= 34.8 THEN 'Obesidade grau 1'
            WHEN BMI >= 34.9 AND BMI <= 39.8 THEN 'Obesidade grau 2'
            ELSE 'Obesidade grau 3'
        END AS Classificação
        FROM diabetes
        )as sub1
      GROUP BY sub1.Classificação
      ORDER BY QtClassif DESC'''

re_d4 = pd.read_sql_query(d4, conn)
re_d4
```

```python
# Quantas pessoas são diabéticas? Quantas não são?

d5 = ''' SELECT Outcome,
          COUNT(*) AS QtOutcome,
          ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM diabetes)) AS Percentual
          FROM diabetes
          GROUP BY Outcome
      '''

re_d5 = pd.read_sql_query(d5, conn)
re_d5
```

- BMI: A maioria dos usuários do dataset possui peso acima da taxa recomendada. Aproximadamente 71% dos usuários estão acima do peso.

# Análise bivariada.

```python
# Quais são as 10 idades que mais registraram casos para diabetes?

f1 = ''' SELECT Age, COUNT(*) As QtUsuDiabetes
          FROM diabetes
          WHERE Outcome = 1
          GROUP BY Age
          ORDER BY QtUsuDiabetes DESC
          LIMIT 10 '''

re_f1 = pd.read_sql_query(f1, conn)
re_f1
```

```python
# Quais são as 10 idades que menos registraram casos para diabetes?

f2 = ''' SELECT Age, COUNT(*) As QtUsuDiabetes
          FROM diabetes
          WHERE Outcome = 1
          GROUP BY Age
          ORDER BY QtUsuDiabetes ASC
          LIMIT 10 '''

re_f2 = pd.read_sql_query(f2, conn)
re_f2
```

```python
# O número de gestações, influencia na obtenção de diabetes?


f3 = '''
      SELECT
    Pregnancies,
    COUNT(*) AS QtPreg,
    SUM(CASE WHEN Outcome = 1 THEN 1 ELSE 0 END) AS QtDiabetes,

    ROUND(SUM(CASE WHEN Outcome = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS Percentual

FROM
    diabetes
GROUP BY
    Pregnancies
ORDER BY Percentual DESC
           '''

re_f3 = pd.read_sql_query(f3, conn)
re_f3


```

```python
# Existe uma correlação entre os níveis de glicose no sangue e a pressão arterial em pacientes diabéticos?

f4 = ''' SELECT
    (COUNT(*) * SUM(Glucose * BloodPressure) - SUM(Glucose) * SUM(BloodPressure)) /
    (SQRT((COUNT(*) * SUM(Glucose * Glucose) - SUM(Glucose) * SUM(Glucose))) *
     SQRT((COUNT(*) * SUM(BloodPressure * BloodPressure) - SUM(BloodPressure) * SUM(BloodPressure)))) AS Correlation
FROM
    diabetes
WHERE
    Outcome = 1
      '''

re_f4 = pd.read_sql_query(f4, conn)
re_f4
```

```python
# Como a espessura da pele (SkinThickness) está relacionada com o índice de massa corporal (IMC) em pacientes diabéticos?

f5 = '''
        SELECT
        (COUNT(*) * SUM(SkinThickness * BMI) - SUM(SkinThickness) * SUM(BMI)) /
        (SQRT((COUNT(*) * SUM(SkinThickness * SkinThickness) - SUM(SkinThickness) * SUM(SkinThickness))) *
        SQRT((COUNT(*) * SUM(BMI * BMI) - SUM(BMI) * SUM(BMI)))) AS Correlation
        FROM diabetes
        WHERE Outcome = 1
        '''

re_f5 = pd.read_sql_query(f5, conn)
re_f5
```

```python
# Quais são os níveis médios de insulina em pacientes diabéticos e como eles variam de acordo com o IMC?


f6 = ''' SELECT sub1.Classificação,
        AVG(sub1.Insulin) AS Média_Insulina
          FROM(
            SELECT BMI, Insulin,
          CASE
            WHEN BMI >= 0 AND BMI <= 18.4 THEN 'Abaixo do peso'
            WHEN BMI >= 18.5 AND BMI <= 24.8 THEN 'Peso normal'
            WHEN BMI >= 24.9 AND BMI <= 29.8 THEN 'Acima do peso'
            WHEN BMI >= 29.9 AND BMI <= 34.8 THEN 'Obesidade grau 1'
            WHEN BMI >= 34.9 AND BMI <= 39.8 THEN 'Obesidade grau 2'
            ELSE 'Obesidade grau 3'
        END AS Classificação
        FROM diabetes
        WHERE Outcome = 1
        ) as sub1
        GROUP BY sub1.Classificação
      '''

re_f6 = pd.read_sql_query(f6, conn)
re_f6
```

```python
# Existe uma relação entre a idade dos pacientes e a espessura da pele (SkinThickness) em pacientes diabéticos?

f7 = '''

SELECT
        (COUNT(*) * SUM(SkinThickness * Age) - SUM(SkinThickness) * SUM(Age)) /
        (SQRT((COUNT(*) * SUM(SkinThickness * SkinThickness) - SUM(SkinThickness) * SUM(SkinThickness))) *
        SQRT((COUNT(*) * SUM(Age * Age) - SUM(Age) * SUM(Age)))) AS Correlation
        FROM diabetes
        WHERE Outcome = 1

'''

re_f7 = pd.read_sql_query(f7, conn)

re_f7
```

```python
# Quais são os padrões de espessura da pele em pacientes diabéticos e não diabéticos?

f8 = '''

SELECT
    CASE
        WHEN Outcome = 1 THEN 'Diabético'
        ELSE 'Não Diabético'
    END AS Categoria,
    AVG(SkinThickness) AS Média,
    MIN(SkinThickness) AS Mínimo,
    MAX(SkinThickness) AS Máximo
FROM
    diabetes
WHERE
    Outcome IN (0, 1)
GROUP BY
    Outcome
'''

re_f8 = pd.read_sql_query(f8, conn)

re_f8
```

```python
# Existe uma correlação entre os níveis de insulina no sangue e a idade dos pacientes diabéticos?

f9 = '''

SELECT
        (COUNT(*) * SUM(Insulin * Age) - SUM(Insulin) * SUM(Age)) /
        (SQRT((COUNT(*) * SUM(Insulin * Insulin) - SUM(Insulin) * SUM(Insulin))) *
        SQRT((COUNT(*) * SUM(Age * Age) - SUM(Age) * SUM(Age)))) AS Correlation
        FROM diabetes
        WHERE Outcome = 1

'''

re_f9 = pd.read_sql_query(f9, conn)
re_f9
```

```python
# Como o índice de massa corporal (IMC) influencia a ocorrência de diabetes em diferentes faixas etárias?

f10 = '''
SELECT
    CASE
        WHEN Age BETWEEN 20 AND 30 THEN '20-30'
        WHEN Age BETWEEN 31 AND 40 THEN '31-40'
        WHEN Age BETWEEN 41 AND 50 THEN '41-50'
        WHEN Age BETWEEN 51 AND 60 THEN '51-60'
        ELSE 'Mais de 60'
    END AS Faixa_Etária,
    CASE
        WHEN Outcome = 1 THEN 'Diabético'
        ELSE 'Não Diabético'
    END AS Categoria,
    AVG(BMI) AS Média_IMC,
    COUNT(*) AS Total_Pacientes
FROM
    diabetes
GROUP BY
    Faixa_Etária, Categoria
ORDER BY
    Faixa_Etária, Categoria
'''


re_f10 = pd.read_sql_query(f10, conn)
re_f10
```

# Conclusões finais a partir das consultas anteriores:

- As idade de 25 e 29 anos foram as idades que mais registraram casos para diabetes.
- As idades 67 e 34 anos foram as idades que menos registraram casos para diabetes.
- Com base nas consultas, nota-se uma tendência clara: quanto maior o número de gestações, maior a prevalência de diabetes. Com 100% das mulheres com mais de 14 gestações apresentando diabetes, essa proporção diminui para 84% com 11 gestações e para 62% com 8 gestações. Portanto, fica evidente que o aumento do número de gestações está associado a um maior risco de desenvolver diabetes.
- Fora constatado que não há uma correlação entre os níveis de glicose no sangue e a pressão arterial de pacientes diabéticos, onde o resultado do coeficiente de Pearson é 0.03.
- Fora observado uma correlação positiva moderada entre a espessura da pele e os indíces de IMC, onde o resultado do coeficiente de Pearson é 0.30. Isso significa que, em geral, à medida que o IMC aumenta, a espessura da pele tende a aumentar também, e vice-versa, mas não de forma muito forte.
- Observa-se um maior nível de insulina no sangue para usuários que possuem obesidade de grau 1, grau 2 e grau 3. Essa conexão entre obesidade e hiperinsulinemia exibe a importância de estratégias de prevenção e manejo da obesidade para informar os riscos associados, como o desenvolvimento de diabetes tipo 2 e outras complicações metabólicas.
- Observou-se que não existe correlação entre a idade dos pacientes e a espessura da pele.
- Nos dados da amostra, usuários diabéticos possuem uma espessura da pele maior comparado a usuários não diabéticos.
- Observou-se que não existe correlação entre a idade dos pacientes e níveis de insulina no sangue.
- Conclui-se que existe uma associação entre diabetes e índice de massa corporal (IMC), independente da faixa etária, com os diabéticos tendendo a ter um IMC maior em comparação com os não diabéticos. Além disso, observa-se uma tendência de redução do IMC à medida que as faixas etárias avançam. Essas informações mostram a importância do controle do peso corporal e da prevenção da obesidade como medidas essenciais na gestão e na prevenção do diabetes, independente da idade.

Isso pode ajudar a identificar indivíduos com maior risco de complicações e direcionar intervenções preventivas específicas.

- Integração de cuidados de saúde: Promover uma abordagem multidisciplinar e integrada no cuidado de pacientes diabéticos, envolvendo médicos de várias especialidades, enfermeiros, nutricionistas, e outros profissionais de saúde. Isso pode melhorar a coordenação do tratamento, a adesão do paciente e os resultados de saúde a longo prazo.
